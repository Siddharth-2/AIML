```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor

# Generate random housing data
np.random.seed(42)
data = pd.DataFrame({
    'area': np.random.randint(1000, 5000, size=100),
    'bedrooms': np.random.randint(1, 6, size=100),
    'age': np.random.randint(5, 100, size=100),
    'price': np.random.randint(100000, 1000000, size=100)
})

# Separate features and target variable
features = data.drop("price", axis=1)
target = data["price"]

# Define the KNN model
model = KNeighborsRegressor(n_neighbors=5)

# Train the model
model.fit(features, target)

# Make predictions on a new data point
new_data = [2500, 3, 25]  # Area: 2500 sq.ft, 3 bedrooms, 25 years old

predicted_price = model.predict([new_data])[0]

# Print the predicted price
print("Predicted price:", predicted_price)
```
